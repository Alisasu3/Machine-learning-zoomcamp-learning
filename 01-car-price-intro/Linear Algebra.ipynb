Vector-Vector Multiplication

```python
def vector_vector_multiplication(u,v):
    assert u.shape[0] == v.shape[0]
    n=u.shape[0]
    result=0.0
    for i in range(n):
        result=result+u[i]*v[i]
    return result
```

```python
vector_vector_multiplication(u,v)
```

```python
u.dot(v) #same result as previous function
```

Matrix-Vector Multiplication

```python
b = np.array([
    [2,4,5,6],
    [1,2,1,2],
    [3,1,2,1]
])
```

```python
b
```

```python
def matrix_vector_multiplication(b,v):
    assert b.shape[1]==v.shape[0]
    num_rows=b.shape[0]
    result=np.zeros(num_rows)
    for i in range(num_rows):
        result[i]=vector_vector_multiplication(b[i],v)
    return result
```

```python
matrix_vector_multiplication(b,v)
```

```python
b.dot(v)
```

Matrix-Matrix Multiplication

```python
m = np.array([
    [1,1,2],
    [0,0.5,1],
    [0,2,1],
    [2,1,0],
])
```

```python
def matrix_matrix_multiplication(b,m):
    assert b.shape[1]==m.shape[0]
    num_rows=b.shape[0]
    num_cols=m.shape[1]
    result=np.zeros((num_rows,num_cols))

    for i in range(num_cols):
        mi=m[:,i]
        bmi=matrix_vector_multiplication(b,mi)
        result[:,i]=bmi
    return result
```

```python
matrix_matrix_multiplication(b,m)
```

```python
b.dot(m)
```
